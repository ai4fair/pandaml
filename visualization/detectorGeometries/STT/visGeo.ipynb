```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
from matplotlib.patches import Circle
import numpy as np

data: pd.DataFrame = pd.read_csv('../../simulationChainMacros/tubePos.csv')
data.columns = data.columns.str.strip()

print(data.head())

xMax: float = data["x"].max()
xMin: float = data["x"].min()

yMax: float = data["y"].max()
yMin: float = data["y"].min()

numStraightTubes: int = 0
numPosSkewedTubes: int = 0
numNegSkewedTubes: int = 0

fig, ax= plt.subplots()

for index, row in data.iterrows():
	if row['angle'] == 0:
		if numStraightTubes == 0:
			straightTubeOuter = Circle((row['x'], row['y']), row['rMin']+row['rMax'], fc='gray', ec=None)
			straightTubeInner = Circle((row['x'], row['y']), row['rMin'], fc='tab:red',ec=None, label = '0°')
		else:
			straightTubeOuter = Circle((row['x'], row['y']), row['rMin']+row['rMax'], fc='gray', ec=None)
			straightTubeInner = Circle((row['x'], row['y']), row['rMin'], fc='tab:red',ec=None)
		ax.add_patch(straightTubeOuter)
		ax.add_patch(straightTubeInner)
		numStraightTubes += 1
	elif np.sign(row['angle']) == 1:
		if numPosSkewedTubes == 0:
			posSkewedTube = Circle((row['x'], row['y']), row['rMin'], color='tab:blue', alpha=0.5, label = '+3°')
		else:
			posSkewedTube = Circle((row['x'], row['y']), row['rMin'], color='tab:blue', alpha=0.5)
		ax.add_patch(posSkewedTube)
		numPosSkewedTubes += 1
	elif np.sign(row['angle']) == -1:
		if numNegSkewedTubes == 0:
			negSkewedTube = Circle((row['x'], row['y']), row['rMin'], color='tab:orange', alpha=0.5, label = '-3°')
		else:
			negSkewedTube = Circle((row['x'], row['y']), row['rMin'], color='tab:orange', alpha=0.5)
		ax.add_patch(negSkewedTube)
		numNegSkewedTubes += 1
	else:
		raise Exception("Invalid angle")
 
ax.set_xlim(xMin-1, xMax+1)
ax.set_ylim(yMin-1, yMax+1)

ax.set_aspect('equal')

plt.legend(loc='upper right')

plt.show()
```
